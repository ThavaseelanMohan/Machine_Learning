```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
scal = MinMaxScaler()
```

```python
df[["Schizophrenia","Bipolar","Eating","Anxiety","Drug use","Depression","Alcohol use"]] = scal.fit_transform(df[["Schizophrenia","Bipolar","Eating","Anxiety","Drug use","Depression","Alcohol use"]])
```

```python
df = df.drop("Code", axis=1)
```

```python
df
```

```python
## Remove Outliers
for i in ['Schizophrenia','Bipolar','Eating','Anxiety','Drug use','Depression','Alcohol use']:
    Q1 = df[i].quantile(0.05)
    Q2 = df[i].quantile(0.90)
    IQR = Q2 - Q1
    lower = Q1 - 3 * IQR
    upper = Q2 +3 * IQR
    df = df[(df[i] >= lower) & (df[i] <= upper)]
df
```

```python
Normalize = df.groupby('Entity').agg({
    'Schizophrenia':'mean',
    'Bipolar':'mean',
    'Eating':'mean',
    'Anxiety':'mean',
    'Drug use':'mean',
    'Depression':'mean',
    'Alcohol use':'mean'
}).reset_index()
```
